# Part 1.Loading the dataset
## Using pandas load the dataset (load remotely, not locally)
## Output the first 15 rows of the data
## Display a summary of the table information (number of datapoints, etc.)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
#read the csv file
dataset = pd.read_csv("https://raw.githubusercontent.com/profmcnich/example_notebook/main/science_data_large.csv")
#print the first 15 rows of data 
dataset.head(15)
```

```python
#print the information about the dataset
dataset.info()
```

# Part 2. Splitting the dataset
## Take the pandas dataset and split it into our features (X) and label (y)

## Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)

```python
#Split the dataset. X = temperature and Mol, y = size
X = dataset[["Temperature 째C","Mols KCL"]]
y = dataset["Size nm^3"]
#Split the dataset into train dataset and test dataset
X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.1,random_state=10)
X_train.shape,y_train.shape
X_test.shape,y_test.shape
```

# Part 3. Perform a Linear Regression
## Use sklearn to train a model on the training set

## Create a sample datapoint and predict the output of that sample with the trained model

## Report on the score for that model, in your own words (markdown, not code) explain what the score means

## Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX

```python
#Linear regression model for training
reg = LinearRegression()

reg.fit(X_train,y_train)
reg.score(X_test,y_test)
```

```python
#the score function will return the coefficient of determination of the prediction. The best possible score is 1.0. In this case the score is 
#0.88 which means the model is accurate for the most of the predictions.
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Part 4. Use Cross Validation
## Use the cross_val_score function to repeat your experiment across many shuffles of the data

## Report on their finding and their significance

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(reg,X,y,cv=12))
```

```python
#The lowest value is 0.8042 and the highest is 0.8926. The average value approximately 0.85 which means 
#the initial score is better than the average score. Also the model can be improved
```

# Part 5. Using Polynomial Regression
## Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2

## Report on the metrics and output the resultant equation as you did in Part 3.

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(2)
pf = poly.fit_transform(X)
```

```python
from sklearn import linear_model
poly_reg = linear_model.Lasso(alpha=0.1)
poly_reg.fit(pf,y)
```

```python
poly_reg.predict(pf)
```

```python
poly_reg.score(pf,y)
```

```python
#The score of the this model is 0.999 which is very close to 1. It means the model is very accurate for predicting.
```
